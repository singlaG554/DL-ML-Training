```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create random dataset
np.random.seed(0)  # For reproducibility
data = np.random.uniform(low=-100, high=100, size=(600, 15))
df = pd.DataFrame(data, columns=[f'Col{i+1}' for i in range(15)])

df.to_csv("random_dataset_q14.csv", index=False) # saves df file to csv
#view file
df = pd.read_csv("random_dataset_q14.csv")
print(df.head(4)) # 4 rows with all columns

# (i) Scatter plot between Column 5 and Column 6
plt.figure(figsize=(8, 6))
plt.scatter(df['Col5'], df['Col6'], color='blue', alpha=0.6)
plt.title('Scatter Plot between Column 5 and Column 6')
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.grid(True)
plt.show()

# (ii) Histogram of each column in single graph
df.plot(kind='hist', bins=30, alpha=0.6, figsize=(12, 6), legend=True)
plt.title('Histograms of All Columns')
plt.xlabel('Value Range')
plt.grid(True)
plt.show()

# (iii) Box plot of each column in single graph
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, orient='h')
plt.title('Box Plot of Each Column')
plt.xlabel('Value')
plt.show()
```
