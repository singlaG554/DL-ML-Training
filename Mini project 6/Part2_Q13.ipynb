```python
# Ques 13
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Create each column range
col1_4 = np.random.uniform(-10, 10, (500, 4))
col5_8 = np.random.uniform(10, 20, (500, 4))
col9_10 = np.random.uniform(-100, 100, (500, 2))

# Combine into one dataset
data = np.hstack((col1_4, col5_8, col9_10))

# Convert to DataFrame
df = pd.DataFrame(data, columns=[f'Col{i}' for i in range(1, 11)])



df.to_csv("random_dataset_q13.csv", index=False) # saves df file to csv
#view file
df = pd.read_csv("random_dataset_q13.csv")
print(df.head(4)) # 4 rows with all columns
```

```python
# Ques 13 (i)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow method
inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df)
    inertia.append(km.inertia_)

# Plot Elbow Graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.title('Elbow Method For Optimal k (K-Means)')
plt.grid(True)
plt.show()
```

```python
# Ques 13 (ii)
import scipy.cluster.hierarchy as sch

# Dendrogram
plt.figure(figsize=(10, 6))
dendrogram = sch.dendrogram(sch.linkage(df, method='ward'))
plt.title('Dendrogram (Hierarchical Clustering)')
plt.xlabel('Data points')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()
```
